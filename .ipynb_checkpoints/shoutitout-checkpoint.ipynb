```python
import json
import datetime, time

json_data = open('/Users/nuwanda/Documents/Projects/shoutitout/republic.json','r')
#d = json_data.readlines()
d=json.load(json_data)

```

```python
#republic
for i in range(0,len(d['items'])):
    #print i
    a = (d['items'][i]['snippet']['title']).encode('ascii', 'ignore').split("|")[0].strip()
    b = filter(None,(d['items'][i]['snippet']['description']).encode('ascii', 'ignore').split("\n"))[1]
    c = datetime.datetime.strptime(d['items'][i]['snippet']['publishedAt'],"%Y-%m-%dT%H:%M:%S.000Z")
    k = datetime.timedelta(hours=6)
    e = (c - k).strftime("%Y-%m-%d")
    print ":::".join([a,b,e])
```
